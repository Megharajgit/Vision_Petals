```python
# ===============================================================
# 🌸 Fine-tune BLIP on Flower Captions Dataset (Stable Version)
# ===============================================================

!pip install transformers datasets accelerate torch torchvision pillow -q

from transformers import BlipProcessor, BlipForConditionalGeneration
from datasets import load_dataset
from torch.utils.data import DataLoader
from tqdm import tqdm
from PIL import Image
import torch
import requests

# ---------------------------------------------------
# 1️⃣ Load model and processor
# ---------------------------------------------------
model_name = "Salesforce/blip-image-captioning-base"
processor = BlipProcessor.from_pretrained(model_name)
model = BlipForConditionalGeneration.from_pretrained(model_name)

device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

# ---------------------------------------------------
# 2️⃣ Load dataset
# ---------------------------------------------------
dataset = load_dataset("AhmadMustafa/flower-captions-blip")

# ---------------------------------------------------
# 3️⃣ Preprocess each example (convert to tensors)
# ---------------------------------------------------
def preprocess_function(example):
    image = example["image"].convert("RGB")
    text = example["text"]
    inputs = processor(
        images=image,
        text=text,
        return_tensors="pt",
        padding="max_length",
        truncation=True
    )
    return {
        "pixel_values": inputs["pixel_values"].squeeze(0),
        "input_ids": inputs["input_ids"].squeeze(0),
        "attention_mask": inputs["attention_mask"].squeeze(0),
    }

train_dataset = dataset["train"].map(preprocess_function, remove_columns=["image", "text"])

# 🔧 Make sure outputs are torch tensors (avoids all stack errors)
train_dataset.set_format(type="torch", columns=["pixel_values", "input_ids", "attention_mask"])

# ---------------------------------------------------
# 4️⃣ Dataloader & Optimizer
# ---------------------------------------------------
train_dataloader = DataLoader(train_dataset, batch_size=2, shuffle=True)
optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)

# ---------------------------------------------------
# 5️⃣ Training Loop
# ---------------------------------------------------
model.train()
epochs = 3

for epoch in range(epochs):
    print(f"\n🌸 Epoch {epoch+1}/{epochs}")
    total_loss = 0
    for batch in tqdm(train_dataloader):
        optimizer.zero_grad()

        pixel_values = batch["pixel_values"].to(device)
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)

        outputs = model(
            pixel_values=pixel_values,
            input_ids=input_ids,
            attention_mask=attention_mask,
            labels=input_ids
        )
        loss = outputs.loss
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    print(f"✅ Average Loss: {total_loss / len(train_dataloader):.4f}")

# ---------------------------------------------------
# 6️⃣ Save the fine-tuned model
# ---------------------------------------------------
model.save_pretrained("./blip-flower-caption-model")
processor.save_pretrained("./blip-flower-caption-model")

print("\n🌼 Model fine-tuned and saved successfully!")

# ---------------------------------------------------
# 7️⃣ Inference: Caption any flower image
# ---------------------------------------------------
def generate_caption(image_url):
    image = Image.open(requests.get(image_url, stream=True).raw).convert('RGB')
    inputs = processor(image, return_tensors="pt").to(device)

    # Increase max_length & use beam search
    out = model.generate(**inputs, max_length=50, num_beams=5)

    caption = processor.decode(out[0], skip_special_tokens=True)
    print(f"🌺 Caption: {caption}")



# ---------------------------------------------------
# 8️⃣ Example: Try captioning a flower
# ---------------------------------------------------
generate_caption("https://storage.googleapis.com/sfr-vision-language-research/BLIP/demo.jpg")

```

```python
!pip install transformers datasets torch pillow -q

from transformers import BlipProcessor, BlipForConditionalGeneration
from datasets import load_dataset
from torch.utils.data import DataLoader
from tqdm import tqdm
from PIL import Image
import torch
import requests
import os

# ---------------------------------------------------
# 0️⃣ Mount Google Drive and set save folder
# ---------------------------------------------------
from google.colab import drive
drive.mount('/content/drive')

# 👉 change this to your folder in Drive
SAVE_DIR = "/content/drive/MyDrive/BLIP"
os.makedirs(SAVE_DIR, exist_ok=True)
print("Models will be saved to:", SAVE_DIR)

# ---------------------------------------------------
# 1️⃣ Load model and processor
# ---------------------------------------------------
model_name = "Salesforce/blip-image-captioning-base"
processor = BlipProcessor.from_pretrained(model_name)
model = BlipForConditionalGeneration.from_pretrained(model_name)

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)
model.to(device)

# ---------------------------------------------------
# 2️⃣ Load ONLY pranked03 dataset
# ---------------------------------------------------
dataset = load_dataset("pranked03/flowers-blip-captions")  # has 'train' split

train_full = dataset["train"]
print(f"Total training examples in pranked03: {len(train_full)}")

# Split into train / val for tracking loss & accuracy
split = train_full.train_test_split(test_size=0.1, seed=42)
train_ds = split["train"]
val_ds   = split["test"]

print(f"Train examples: {len(train_ds)}")
print(f"Val examples:   {len(val_ds)}")

# ---------------------------------------------------
# 3️⃣ Preprocess each example (same style as your original code)
# ---------------------------------------------------
def preprocess_function(example):
    # Ensure image is RGB
    image = example["image"].convert("RGB")
    text = example["text"]

    inputs = processor(
        images=image,
        text=text,
        return_tensors="pt",
        padding="max_length",
        truncation=True
    )

    # labels = input_ids, we will use attention_mask to compute accuracy
    return {
        "pixel_values": inputs["pixel_values"].squeeze(0),
        "input_ids": inputs["input_ids"].squeeze(0),
        "attention_mask": inputs["attention_mask"].squeeze(0),
    }

train_dataset = train_ds.map(preprocess_function, remove_columns=["image", "text"])
val_dataset   = val_ds.map(preprocess_function,   remove_columns=["image", "text"])

# Set format to torch tensors (IMPORTANT: same as your working code)
train_dataset.set_format(type="torch", columns=["pixel_values", "input_ids", "attention_mask"])
val_dataset.set_format(type="torch",   columns=["pixel_values", "input_ids", "attention_mask"])

# ---------------------------------------------------
# 4️⃣ Dataloaders & Optimizer
# ---------------------------------------------------
train_dataloader = DataLoader(train_dataset, batch_size=2, shuffle=True)
val_dataloader   = DataLoader(val_dataset,   batch_size=2, shuffle=False)

optimizer = torch.optim.AdamW(model.parameters(), lr=2e-5)  # smaller LR to reduce repetition

# ---------------------------------------------------
# 5️⃣ Training Loop with accuracy + best-model checkpoint
# ---------------------------------------------------
model.train()
epochs = 3

best_accuracy = 0.0  # track best val accuracy
best_model_dir = os.path.join(SAVE_DIR, "best_model_pranked")
os.makedirs(best_model_dir, exist_ok=True)

for epoch in range(epochs):
    print(f"\n Epoch {epoch+1}/{epochs}")
    total_loss = 0.0

    # training
    model.train()
    for batch in tqdm(train_dataloader):
        optimizer.zero_grad()

        pixel_values = batch["pixel_values"].to(device)
        input_ids    = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)

        outputs = model(
            pixel_values=pixel_values,
            input_ids=input_ids,
            attention_mask=attention_mask,
            labels=input_ids
        )
        loss = outputs.loss
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    avg_train_loss = total_loss / len(train_dataloader)
    print(f"✅ Average Train Loss: {avg_train_loss:.4f}")

    # -------------------------------
    # Validation: loss + token accuracy
    # -------------------------------
    model.eval()
    val_loss = 0.0
    correct_tokens = 0
    total_tokens   = 0

    with torch.no_grad():
        for batch in val_dataloader:
            pixel_values = batch["pixel_values"].to(device)
            input_ids    = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)

            outputs = model(
                pixel_values=pixel_values,
                input_ids=input_ids,
                attention_mask=attention_mask,
                labels=input_ids
            )

            loss = outputs.loss
            val_loss += loss.item()

            # token-level accuracy (where attention_mask == 1)
            logits = outputs.logits  # (batch, seq_len, vocab)
            preds = logits.argmax(dim=-1)
            mask = attention_mask.bool()

            correct_tokens += (preds[mask] == input_ids[mask]).sum().item()
            total_tokens   += mask.sum().item()

    avg_val_loss = val_loss / len(val_dataloader)
    val_accuracy = correct_tokens / total_tokens if total_tokens > 0 else 0.0

    print(f"🎯 Val Loss: {avg_val_loss:.4f}")
    print(f"🎯 Val Token Accuracy: {val_accuracy * 100:.2f}%")

    # Save best model based on val accuracy
    if val_accuracy > best_accuracy:
        best_accuracy = val_accuracy
        model.save_pretrained(best_model_dir)
        processor.save_pretrained(best_model_dir)
        print(f"🏅 New BEST model saved at: {best_model_dir} (acc={best_accuracy*100:.2f}%)")

# ---------------------------------------------------
# 6️⃣ Also save the final (last epoch) model
# ---------------------------------------------------
final_model_dir = os.path.join(SAVE_DIR, "pranked-flower-last-epoch")
os.makedirs(final_model_dir, exist_ok=True)
model.save_pretrained(final_model_dir)
processor.save_pretrained(final_model_dir)

print("\n🌼 Training finished!")
print("📌 Best model dir: ", best_model_dir)
print("📌 Last epoch dir:", final_model_dir)

# ---------------------------------------------------
# 7️⃣ Inference: Caption any flower image with current (last-epoch) model
# ---------------------------------------------------
def generate_caption(image_url):
    image = Image.open(requests.get(image_url, stream=True).raw).convert('RGB')
    inputs = processor(image, return_tensors="pt").to(device)

    out = model.generate(
        **inputs,
        max_new_tokens=20,       # short caption like dataset
        num_beams=5,
        no_repeat_ngram_size=3,  # avoid "is is is is"
        repetition_penalty=2.0,
    )
    caption = processor.decode(out[0], skip_special_tokens=True)
    print(f"🌺 Caption: {caption}")

# Test once after training
generate_caption("https://storage.googleapis.com/sfr-vision-language-research/BLIP/demo.jpg")
```

```python
import os
from transformers import BlipProcessor, BlipForConditionalGeneration, pipeline
from PIL import Image
import torch
import gradio as gr
from google.colab import drive

# 1. Setup
drive.mount('/content/drive', force_remount=True)
MODEL_DIR = "/content/drive/MyDrive/BLIP/best_model_pranked"

# 2. Load Models
print("Loading Captioning Model...")
processor = BlipProcessor.from_pretrained(MODEL_DIR)
model = BlipForConditionalGeneration.from_pretrained(MODEL_DIR)

device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
model.eval()

print("Loading Classifier...")
classifier = pipeline("zero-shot-image-classification", model="openai/clip-vit-base-patch32", device=device)

def caption_image(image):
    if image is None:
        return "not a flower image"

    image_rgb = image.convert("RGB")

    # --- SIMPLIFIED FILTER ---
    # Instead of hard-coding 20 items, we use two distinct concepts.
    # "Concept" vs "Noise/Other"
    labels_to_check = ["a flower", "random object"]

    # Run classification
    results = classifier(image_rgb, candidate_labels=labels_to_check)

    # Get the top prediction and its CONFIDENCE SCORE (0 to 1)
    top_prediction = results[0]['label']
    confidence_score = results[0]['score']

    print(f"Prediction: {top_prediction}, Confidence: {confidence_score:.4f}")

    # --- THRESHOLD CHECK ---
    # 1. It must predict "a flower"
    # 2. It must be VERY confident (e.g., > 0.9 or 90%)
    # If it's a pink ball, the score for "flower" might be 0.6, so this will reject it.
    if top_prediction != "a flower" or confidence_score < 0.90:
        return "not a flower image"

    # --- CAPTIONING ---
    inputs = processor(image_rgb, return_tensors="pt").to(device)

    with torch.no_grad():
        out = model.generate(
            **inputs,
            max_new_tokens=20,
            num_beams=5,
            no_repeat_ngram_size=3,
            repetition_penalty=2.0,
        )

    caption = processor.decode(out[0], skip_special_tokens=True)
    return caption

# 3. Launch App
demo = gr.Interface(
    fn=caption_image,
    inputs=gr.Image(type="pil", label="Upload Image"),
    outputs=gr.Textbox(label="Result"),
    title="Flower Captioner",
    # description="This tool generates captions only if it is >90% sure the image is a flower.",
)

demo.launch(share=True)
```

```python
# Install necessary library (faiss-cpu)
!pip install faiss-cpu safetensors -q

import faiss
import numpy as np
import torch
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration
from datasets import load_dataset
from tqdm import tqdm
import os
import json
import shutil # Used for copying files
from google.colab import drive

# --- CRITICAL STEP: Mount Drive ---
# Execute this in a separate cell if not already done, or ensure it runs before this block.
# drive.mount('/content/drive')

# --- CONFIGURATION: Project Paths ---
GD_MODEL_DIR = "/content/drive/MyDrive/BLIP/best_model_pranked" # Fine-tuned BLIP model path
PROJECT_DIR = "/content/drive/MyDrive/BLIP"

# Define the folder for index files and metadata
FAISS_DIR = os.path.join(PROJECT_DIR, "faiss_index")
INDEX_FILENAME = "image_vectors.index"
METADATA_FILENAME = "caption_metadata.json"

# Create the persistence directory if it doesn't exist
os.makedirs(FAISS_DIR, exist_ok=True)
print(f"Persistence directory created/verified: {FAISS_DIR}")
# ---------------------------------------------------

# Load BLIP Model and Processor
print("\nLoading BLIP Model and Processor...")

# FIX 1: Load processor from the local directory first (This is usually safer than using the remote 'Salesforce/...')
processor = BlipProcessor.from_pretrained(GD_MODEL_DIR, local_files_only=True)

# FIX 2: Load model from the local directory, forcing local files.
model = BlipForConditionalGeneration.from_pretrained(GD_MODEL_DIR, local_files_only=True)

device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
model.eval()

# Load the dataset (all training examples)
dataset = load_dataset("pranked03/flowers-blip-captions")
train_dataset = dataset["train"]
print(f"Loaded {len(train_dataset)} examples for indexing.")

# 1. GENERATE EMBEDDINGS AND COLLECT METADATA
image_embeddings = []
captions_metadata = [] # List to store the actual caption text for look-up

print("Generating embeddings for all images in the dataset...")

for i in tqdm(range(len(train_dataset)), desc="Generating Embeddings"):
    # Get image and caption
    item = train_dataset[i]
    image = item["image"].convert("RGB")
    caption = item["text"]

    # Process image to get inputs
    pixel_values = processor(images=image, return_tensors="pt").pixel_values.to(device)

    # Extract features using the model's vision encoder
    with torch.no_grad():
        vision_outputs = model.vision_model(pixel_values=pixel_values)
        features = vision_outputs.pooler_output # The image vector embedding

    # Store
    embedding = features.cpu().numpy()
    image_embeddings.append(embedding)
    captions_metadata.append(caption) # Store the text corresponding to this image index

# 2. BUILD THE FAISS INDEX
embedding_matrix = np.vstack(image_embeddings)
dimension = embedding_matrix.shape[1]

# Create the Index (L2 distance is standard for unnormalized embeddings)
index = faiss.IndexFlatL2(dimension)
index.add(embedding_matrix)

print(f"✅ FAISS Index Built! Total vectors stored: {index.ntotal} (Dimension: {dimension})")

# 3. PERSISTENCE: SAVE INDEX AND METADATA TO DRIVE
try:
    # Save Index File
    faiss.write_index(index, os.path.join(FAISS_DIR, INDEX_FILENAME))

    # Save Metadata (Captions) File
    with open(os.path.join(FAISS_DIR, METADATA_FILENAME), 'w') as f:
        json.dump(captions_metadata, f)

    print(f"\n✨ SUCCESS: Vector database saved permanently to {FAISS_DIR}")

except Exception as e:
    print(f"\n❌ ERROR during saving: {e}")
```
